```python
from sklearn.metrics.pairwise import cosine_distances
```

```python
def compute_cosine_distances(df, word1_embeddings_1800, word1_embeddings_1820, word1_embeddings_1840, word1_embeddings_1860, word1_embeddings_1880, word1_embeddings_1900, word1_embeddings_1920, word1_embeddings_1940, word1_embeddings_1960, word1_embeddings_1980, word1_embeddings_1990, word2_embeddings_1800, word2_embeddings_1820, word2_embeddings_1840, word2_embeddings_1860, word2_embeddings_1880, word2_embeddings_1900, word2_embeddings_1920, word2_embeddings_1940, word2_embeddings_1960, word2_embeddings_1980, word2_embeddings_1990):
    # Lists to store cosine distances
    cosine_distances_1800 = []
    cosine_distances_1820 = []
    cosine_distances_1840 = []
    cosine_distances_1860 = []
    cosine_distances_1880 = []
    cosine_distances_1900 = []
    cosine_distances_1920 = []
    cosine_distances_1940 = []
    cosine_distances_1960 = []
    cosine_distances_1980 = []
    cosine_distances_1990 = []

    for index, row in df.iterrows():
        word1_word_1800_embedding = word1_embeddings_1800.get(row['word 1']).reshape(1, -1)
        word1_word_1820_embedding = word1_embeddings_1820.get(row['word 1']).reshape(1, -1)
        word1_word_1840_embedding = word1_embeddings_1840.get(row['word 1']).reshape(1, -1)
        word1_word_1860_embedding = word1_embeddings_1860.get(row['word 1']).reshape(1, -1)
        word1_word_1880_embedding = word1_embeddings_1880.get(row['word 1']).reshape(1, -1)
        word1_word_1900_embedding = word1_embeddings_1900.get(row['word 1']).reshape(1, -1)
        word1_word_1920_embedding = word1_embeddings_1920.get(row['word 1']).reshape(1, -1)
        word1_word_1940_embedding = word1_embeddings_1940.get(row['word 1']).reshape(1, -1)
        word1_word_1960_embedding = word1_embeddings_1960.get(row['word 1']).reshape(1, -1)
        word1_word_1980_embedding = word1_embeddings_1980.get(row['word 1']).reshape(1, -1)
        word1_word_1990_embedding = word1_embeddings_1990.get(row['word 1']).reshape(1, -1)

        word2_word_1800_embedding = word2_embeddings_1800.get(row['word 2']).reshape(1, -1)
        word2_word_1820_embedding = word2_embeddings_1820.get(row['word 2']).reshape(1, -1)
        word2_word_1840_embedding = word2_embeddings_1840.get(row['word 2']).reshape(1, -1)
        word2_word_1860_embedding = word2_embeddings_1860.get(row['word 2']).reshape(1, -1)
        word2_word_1880_embedding = word2_embeddings_1880.get(row['word 2']).reshape(1, -1)
        word2_word_1900_embedding = word2_embeddings_1900.get(row['word 2']).reshape(1, -1)
        word2_word_1920_embedding = word2_embeddings_1920.get(row['word 2']).reshape(1, -1)
        word2_word_1940_embedding = word2_embeddings_1940.get(row['word 2']).reshape(1, -1)
        word2_word_1960_embedding = word2_embeddings_1960.get(row['word 2']).reshape(1, -1)
        word2_word_1980_embedding = word2_embeddings_1980.get(row['word 2']).reshape(1, -1)
        word2_word_1990_embedding = word2_embeddings_1990.get(row['word 2']).reshape(1, -1)


        cosine_distance_1800 = cosine_distances(word1_word_1800_embedding, word2_word_1800_embedding)
        cosine_distance_1820 = cosine_distances(word1_word_1820_embedding, word2_word_1820_embedding)
        cosine_distance_1840 = cosine_distances(word1_word_1840_embedding, word2_word_1840_embedding)
        cosine_distance_1860 = cosine_distances(word1_word_1860_embedding, word2_word_1860_embedding)
        cosine_distance_1880 = cosine_distances(word1_word_1880_embedding, word2_word_1880_embedding)
        cosine_distance_1900 = cosine_distances(word1_word_1900_embedding, word2_word_1900_embedding)
        cosine_distance_1920 = cosine_distances(word1_word_1920_embedding, word2_word_1920_embedding)
        cosine_distance_1940 = cosine_distances(word1_word_1940_embedding, word2_word_1940_embedding)
        cosine_distance_1960 = cosine_distances(word1_word_1960_embedding, word2_word_1960_embedding)
        cosine_distance_1980 = cosine_distances(word1_word_1980_embedding, word2_word_1980_embedding)
        cosine_distance_1990 = cosine_distances(word1_word_1990_embedding, word2_word_1990_embedding)

        cosine_distances_1800.append(cosine_distance_1800[0][0])
        cosine_distances_1820.append(cosine_distance_1820[0][0])
        cosine_distances_1840.append(cosine_distance_1840[0][0])
        cosine_distances_1860.append(cosine_distance_1860[0][0])
        cosine_distances_1880.append(cosine_distance_1880[0][0])
        cosine_distances_1900.append(cosine_distance_1900[0][0])
        cosine_distances_1920.append(cosine_distance_1920[0][0])
        cosine_distances_1940.append(cosine_distance_1940[0][0])
        cosine_distances_1960.append(cosine_distance_1960[0][0])
        cosine_distances_1980.append(cosine_distance_1980[0][0])
        cosine_distances_1990.append(cosine_distance_1990[0][0])

    # Add the cosine distances as new columns to the dataframe
    df['1800_cosine_distance'] = cosine_distances_1800
    df['1820_cosine_distance'] = cosine_distances_1820
    df['1840_cosine_distance'] = cosine_distances_1840
    df['1860_cosine_distance'] = cosine_distances_1860
    df['1880_cosine_distance'] = cosine_distances_1880
    df['1900_cosine_distance'] = cosine_distances_1900
    df['1920_cosine_distance'] = cosine_distances_1920
    df['1940_cosine_distance'] = cosine_distances_1940
    df['1960_cosine_distance'] = cosine_distances_1960
    df['1980_cosine_distance'] = cosine_distances_1980
    df['1990_cosine_distance'] = cosine_distances_1990

    return df
```
